```python
def Fibonacci_recursive(n):
  if n <= 0:
    return 0
  elif n == 1:
    return 1
  else:
      return Fibonacci_recursive(n-1) + Fibonacci_recursive(n-2)
```

```python
Fibonacci_recursive(6)
```

```python
Fibonacci_recursive(30)
```

```python
def Fibonacci_iter(n):
  if n <= 0:
    return 0
  elif n == 1:
    return 1
  prev = 0
  current = 1
  for _ in range(2, n+1):
    next_value = prev + current
    prev = current
    current = next_value
  return current
```

```python
Fibonacci_iter(6)
```

```python
Fibonacci_iter(70)
```
